```python
df = pd.read_csv("Wounding Summary-set2.csv") # Read in files
idx = df.score >= 6  # Rows in which the score is greater than 6 - save those
df = df[idx]
start = df["pos"] - 10000 # Calculate the left border of LD window
df['start'] = start # Add a column of the left border position
end = df["pos"] + 10000 # Calculate the right border of LD window
df['end'] = end # Add a column of the right border position
output_file_name =  "Wounding Summary-set2-pscore6.csv"
df.to_csv(output_file_name) # Export new generated files using a new file name
```

```python
df1 = pd.read_csv("Reference gene list.csv") # Read the reference gene information file containing all genes in Arabidopsis
df1
```

```python
df2 = pd.read_csv("Wounding Summary-set2-pscore6.csv") # Read the new generated GWAS summary file with start and end position of LD window for each SNP
df2
```

```python
LD = [] #Generate an empty list
array = [] #Generate an empty array
for i in range(len(df2.index)): #Pick up each SNP 
    for j in range(len(df1.index)): #With each SNP, loop up all nearby genes within LD window
        if df1.loc[j,"Chromosome"] == df2.loc[i,"chr"] and df2.loc[i,'end']>=df1.loc[j,'Start'] and df2.loc[i,'start']<=df1.loc[j,'End']:
            gene_ID = df1.loc[j,'Gene']
            LD.append(gene_ID) #Generate a candidate list for each SNP
    array.append(str(LD)) #Generate a candidate array for all significant SNPs
    LD = []
df2['Gene'] = array
df2['Gene'] #Add a new column with candidate gene list
```

```python
df2.to_csv("Wounding-set2-pscore6-candidate gene list.csv") # Export new generated files using a new file name
```
